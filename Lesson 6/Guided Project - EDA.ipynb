````python
# Import the mandatory library
import pandas as pd
# keep the name of dataset into file variable
file = 'imd_notas.xlsx'
# Load the dataset for a pandas.io.excel.ExcelFile object
excel = pd.ExcelFile(file)
# Print the seet names of dataset
print(excel.sheet_names)
# Transform the dataset from an Excel sheet to a Dataframe
data = excel.parse(0)
# Print the first five rows of dataset
data.head()
````

````python
#4 - How many students have been registered?
len(data['a_ID'].unique())
````

````python
#5 What is the period in years of the records in the database?
data['ano_disciplina'].unique()
````

````python
#6 How many disciplines were analyzed?
data['disciplina_ID'].unique()
````

````python
#7 List the total number of students present in the first five ZIP codes with the most records.
data[~data['a_ID'].duplicated()]['CEP'].value_counts()[0:5].sum()
````

````python
#8 Calculate the average grades for each discipline
import numpy as np
data.pivot_table(index=["disciplina_ID"], values="nota", aggfunc=np.mean)
````

````python
#9 Calculate the average grades for each discipline grouped by year.
bkp_data = data.pivot_table(index=["disciplina_ID", "ano_disciplina"], values="nota", aggfunc=np.mean)
bkp_data

# bonus
# bkp_data.reset_index().sort_values(['disciplina_ID','nota'], 
#                              ascending=[True, False]).set_index(['disciplina_ID','ano_disciplina'])
````

````python
#10 Calculate the number of students approved and failed by discipline and grouped per year
bkp_data = data.pivot_table(index=["disciplina_ID", 'ano_disciplina','status.disciplina'], 
                            values="nota", aggfunc='count')
bkp_data

# Bonus
#print(bkp_data.axes)
#print(bkp_data.index.get_level_values(1))
#bkp_data.loc[(bkp_data.index.get_level_values(1) == 2014) &
#            (bkp_data.index.get_level_values(2) == 'Aprovado')]
````
